### Delež dnevnega trgovanja "največjih" kriptovalut

Naslednji graf prikazuje količino dnevnega trgovanja Bitcoin-a, Ethereum-a (oz. Ether-ja), Litecoin-a in Ripple-a, v primerjavi s celotnim trgom.

## Gibanje maksimalnih dnevnih cen

Za analiziranje trendov cen različnih kriptovalut se uporablja vec statističnih metod. Ker imamo opravka s časovnimi vrstami, je glajenje vrste primerna metoda za analizo, saj vsebuje vse bistvene vzorce, ne vsebuje pa šuma in drugih mikrostruktur v podatkih. V zglajeni vrsti lažje odkrijemo dinamiko pojava in jo lahko uporabimo za napovedovanje. 

Uporabili bomo dve vrsti glajenja - glajenje z drsečim povprečjem in eksponentno glajenje.

#### Glajenje z drsečim povprečjem

Glajenje z drsečim povprečjem reda k časovni vrsti priredi glajene vrednosti, ki so povprečja zadnjih k vrednosti.
V ${\color{blue} {analyzing\_data.py}}$ izračunamo glajene vrednosti in jih shranimo v csv. Glajene vrednosti računamo glede na maksimalne dnevne vrednosti za reda $k = 5$ in $k = 14$. Pri manjšem $k$-ju so lahko enkratni skoki bolj izraziti, a lažje ujamemo trend, ki se pojavlja na tedenski ravni. Ko $k$ večamo, zgladimo posamezne šoke in omogočimo analizo dinamike na mesečni oz večmesečni ravni.

###### k = 5:

```python
glajene_vrednosti5 = pd.read_csv('podatki/glajene_vrednosti_drsece_povprecje_high_5.csv', squeeze=True, parse_dates = ["Date"], index_col = 'Date')
#vrednosti5 = glajene_vrednosti5.groupby(['Name'])['GlajeneVrednosti']
pd.pivot_table(data = glajene_vrednosti5, values = "GlajeneVrednosti", columns = "Name", index = "Date").plot(legend = False)
```

###### k = 14:

```python
glajene_vrednosti14 = pd.read_csv('podatki/glajene_vrednosti_drsece_povprecje_high_14.csv', squeeze=True, parse_dates = ["Date"], index_col = 'Date')
# vrednosti = glajene_vrednosti.groupby(['Name'])['GlajeneVrednosti']
pd.pivot_table(data = glajene_vrednosti14, values = "GlajeneVrednosti", columns = "Name", index = "Date").plot(legend = False)
```

Na obeh grafih vidimo, da je trg kriptovalut s cenami posamzenih kriptovalut dosegel vrhunec, prav tako pa grafa opozarjata na predviden še večji padec cen v januarju.  Kot lahko vidimo na grafu, ima večino kriptovalut podoben trend in posledično se na daljši rok večino kriptovalut obnaša okvirno tako, kot se obnaša trg sam. Veliko kriptovalut se trguje za Bitcoin ali Ether in ne neposredno za fiat valuto, kot je naprimer dolar. To pomeni, da je njihova cena zelo odvisna od cene teh dveh, saj v primeru šokov večjih kriptovalut zanihajo tudi cene teh manjših, odvisnih kriptovalut.

#### Eksponentno glajenje

Eksponentno glajenje s parametrom $\alpha$ časovni vrsti priredi glajene vrednosti z rekurzivno formulo.
V ${\color{blue} {analyzing\_data.py}}$ izračunamo glajene vrednosti in jih shranimo v csv. Glajene vrednosti računamo glede na maksimalne dnevne vrednosti za $\alpha = 0.5$.

Rekurzivna formula (vrednosti $l_i$ so glajene vrednosti, vrednosti $y_i$ pa originalni podatki):
$$l_1 = y_1$$
$$l_t = \alpha y_t + (1- \alpha)l_{t-1}$$

```python
glajene_vrednosti05 = pd.read_csv('podatki/glajene_vrednosti_eksponentno_05.csv', squeeze=True, parse_dates = ["Date"], index_col = 'Date')
#vrednosti5 = glajene_vrednosti5.groupby(['Name'])['GlajeneVrednosti']
pd.pivot_table(data = glajene_vrednosti05, values = "GlajeneVrednosti", columns = "Name", index = "Date").plot(legend = False)
```

Kot vidimo na tem grafu, se oblika krivulje oz. ekstremnih vrednosti ponavlja pri večih kriptovalutah, le da dosegajo druge vrednosti. Tu lahko opazimo obliko krivulje, ki je značilna za "borzne balončke". Vrednost neke kriptovalute zelo hitro naraste zaradi nekega razloga, lahko je to zaradi dejanske spremembe notranje vrednosti valute ali pa le zaradi "pump and dump" sheme. Nato jo zaradi te rasti začnejo kupovati tudi vlagatelji, ki je ne kupujejo zaradi notranje vrednosti, ampak zaradi trenda rasti. Na neki točki se ta rast umiri in vlagatelji, ki so v valuto vlagali na začetku, jo prodajo. To sproži padec cene, kateremu sledi še večje prodajanje in nižanje cen. Ker pa nekateri vlagatelji vseeno verjamejo, da je to le začasen padec, ki mu bo sledila še višja rast, se odločijo kupiti sedaj, ko je cena nižja in to povzroči ponovno rast, ki pa večinoma ne doseže prejšnje vrednosti, sledi pa ji še enkratni padec cen, saj je notranja vrednost še vedno nižja kot trenutna cena na trgu. Tej "jami" se reče "bull trap". Kot je lepo razvidno na nekaterih izmed zgornjih grafov, se to lahko ponovi večkrat na krajše ali daljše obdobje.   
Taka oblika krivulje se lahko oz. se pojavi tudi pri padcu trga, ki ni povzročen namerno, kot lahko vidimo na koncu decembra.

V primeru, da gre za "pump and dump" shemo, bo ta najučinkoviteje delovala na tistih kriptovalutah, ki imajo nizko dnevno količino trgovanja in so občutljive na zmerno velike nakupe in prodaje le te. To lahko vidimo tudi na zgornjem grafu. Če se spomnimo, so tiste kriptovalute, poleg Bitcoina, ki so imele najvišje cene kovancev, imele tudi manjšo količino trgovanja. Tu je seveda treba poudariti, da je izvajanje in zavestno sodelovanje pri takih shemah kaznivo.

### Gibanje štirih najbolj razširjenih kriptovalut

Iz zgornjih grafov, kjer smo gledali gibanje cen celotnega trga, zaradi količine podatkov ni razvidno, kako se obnašajo najbolj trgovane kriptovalute. Zato si lahko ogledamo glajena povprečja (vsa glajenja ene posamične kriptovalute predstavimo na enem grafu). Iz glajenj bi lahko ugotovili npr. ciklično gibanje posamezne kriptovalute.

## Korelacijska matrika 

Zelo zanimivo bi bilo vedeti, kakšna je povezava med kriptovalutami - kako so med sabo kolerirane, ali vzpon/padec ene kriptovalute vpliva na gibanje drugih kriptovalut (in če, katerih).
Za pretekle podatke lahko naredimo korelacijsko matriko, ki nam prikaže korelacije, izračunane na podlagi lanskega leta. 

Ali so, in če, kako so spremembe cen različnih kriptovalut med sabo povezane, je težko reči. Imajo pa največji vpliv na celotni trg kriptovalut seveda tiste, s katerimi se največ trguje.

Če želimo izračunane kovariance med poljubnim parom kriptovalut, jih samo dodamo v $izbira\_kriptovalut$.

```python
izbira_kriptovalut = ["Bitcoin", "Ethereum", "Ripple", "Litecoin"]
```

Naredili bomo dve matriki - ena s kovarinacami največjih dnevnih cen kriptovalut in eno s kovariancami najmanjših cen vrednosti kriptovalut.

Za posamezen dan izračunamo procentualne spremembe glede na prejšnji dan. To naredimo za vsako kriptovaluto in vrednosti shranimo v tabelo. 

```python
spremembe_high = [volumen[volumen["Name"] == name]["High"].sort_index().pct_change() for name in izbira_kriptovalut]
spremembe_high = pd.concat(spremembe_high, axis = 1)
spremembe_high.columns = izbira_kriptovalut
korelacija_high = spremembe_high.corr()
korelacija_high
```

```python
spremembe_low = [volumen[volumen["Name"] == name]["Low"].sort_index().pct_change() for name in izbira_kriptovalut]
spremembe_low = pd.concat(spremembe_low, axis = 1)
spremembe_low.columns = izbira_kriptovalut
korelacija_low = spremembe_low.corr()
korelacija_low
```

Kot vidimo v zgornjih korelacijskih matrikah, je gibanje cen korelirano, a manj, kot bi pričakovali. To je lahko posledica dejanske majhne koreliranosti, ali pa dejstva, da smo mi za primerjavo vzeli ekstremne vrednosti vsakega dneva. Če bi imeli več podatkov, naprimer urne, ali pa povprečno trgovalno ceno v dnevu, bi korelacija bila verjetno večja.  
Ampak tudi tu lahko opazimo, da sta Bitcoin in Ether kar korelirana.

Spodaj bomo naredili še korelacijsko matriko povprečja največje, najmanjše in closing vrednosti.

```python
spremembe_hlc = [volumen[volumen["Name"] == name][["High", "Low", "Close"]].mean(axis = 1).sort_index().pct_change() for name in izbira_kriptovalut]
spremembe_hlc = pd.concat(spremembe_hlc, axis = 1)
spremembe_hlc.columns = izbira_kriptovalut
korelacija_hlc = spremembe_hlc.corr()
korelacija_hlc
```

Kot lahko vidimo tudi tu, je Litecoin približno isto koreliran z Bitcoin-om in Ether-jem. Ripple je kar neodvisen od Bitcoina, kar smo lahko opazili tudi na zgornjih grafih, ko je Bitcoin na koncu leta padel, Ripple pa še ne.  

Predvidevamo lahko tudi, da je korelacija tu manjša, ker smo vzemali natančnejše vrednosti sprememb. Če bi jih zaokrožili le na 10% natančno (se pravi 0; 0,1; 0,2; ...), ali pa še bolj grobo, kot naprimer le na kazalce vzpona ali padca kriptovalute, bi morali dobiti višjo koreliranost.

Trg s kriptovalutami je zelo zanimiv, kljub razširjenosti, ki je dokaj očitno v preteklem letu zelo narastla, pa je to nenapovedljiv trg. V mesecu januarju leta 2018 je trg doživel ogromen padec in cene vseh kriptovalut (znanih in manj znanih) so padle tudi za 70%. 

Investiranje v kriptovalute je lahko kakor odlična naložba, tudi ogromna izguba, zato je pri investiranju potrebno poskrbeti za razpršenost (denar vložimo v več različnih kriptovalut).
